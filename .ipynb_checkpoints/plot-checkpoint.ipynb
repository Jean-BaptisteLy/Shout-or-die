```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
sns.set()
```

```python
df0 = pd.read_csv("./Data/loudnessDataN0.csv")
df1 = pd.read_csv("./Data/loudnessDataN1.csv")
df2 = pd.read_csv("./Data/loudnessDataN2.csv")
df3 = pd.read_csv("./Data/loudnessDataN3.csv")
dfs = [df0, df1, df2, df3]
thresholds = pd.read_csv("./Data/thresh.csv")
```

```python
moveThresh, jumpThresh = thresholds["moveThresh"][0], thresholds["jumpThresh"][0]
print(moveThresh, jumpThresh)
```

```python
for i in range(len(dfs)):
    plt.plot(dfs[i]["Time"],dfs[i][" Loudness"], label = "Level "+f"{i}")
    plt.plot(dfs[i]["Time"], [moveThresh]*len(dfs[i]["Time"]), linestyle='--', label = "Move threshold")
    plt.plot(dfs[i]["Time"], [jumpThresh]*len(dfs[i]["Time"]), linestyle='--', label = "Jump threshold")
    plt.xlabel('Time')
    plt.ylabel('Loudness')
    plt.title("Loudness values over time for level " + str(i))
    plt.legend()
    plt.show()
```
